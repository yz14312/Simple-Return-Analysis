```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
TSLA = wb.DataReader("TSLA", data_source = "yahoo", start = "2003-01-01")
TSLA
```

```python
tickers = ["NFLX", "TSLA", "AMZN", "ZM"]

tech_data = pd.DataFrame()

for t in tickers:
    tech_data[t] = wb.DataReader(t, data_source = "yahoo", start = "2020-01-01")["Adj Close"]
```

```python
tech_data
```

```python
tech_data["NFLX"].plot(figsize = (8,5), color = "black")
tech_data["TSLA"].plot(figsize = (8,5))
tech_data["AMZN"].plot(figsize = (8,5))
tech_data["ZM"].plot(figsize = (8,5))

plt.title("Stock Price")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Stock Price ($)")
plt.savefig("F:/Data Science practice/Blog/Tech_data_plot.png")
plt.show()
```

Netflix Stock Simple Returns

```python
NFLX = wb.DataReader("NFLX", data_source = "yahoo", start = "2020-01-01")

NFLX["Simple Return"] = (NFLX["Adj Close"] / NFLX["Adj Close"].shift(1))-1
print(NFLX["Simple Return"])

NFLX["Simple Return"].plot(figsize = (8,5), color = "black")
plt.title("Netflix Simple Return Graph")
plt.xlabel("Time")
plt.ylabel("Simple Return")
plt.savefig("F:/Data Science practice/Blog/Netflix_simple_return.png")
plt.show()
```

```python
avg_return_day = NFLX["Simple Return"].mean()
avg_return_year = avg_return_day *250
print(str(round(avg_return_year, 4) * 100), "%")
```
